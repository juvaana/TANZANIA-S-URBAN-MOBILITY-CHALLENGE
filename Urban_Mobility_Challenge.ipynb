```python
import pandas as pd
import numpy as np
from ydata_profiling import ProfileReport


train=pd.read_csv('train.csv')
test=pd.read_csv('test.csv')


# Automatic EDA process for our project
#This code provides you with an interactive page for all the EDA process from Shapes of the data to the correlation
Profile = ProfileReport(train)



```

```python
#Feature Engineering
#Since there was no null values in the train & test datasets we are going to Encode the categorical values

#1.List the categorical value columns
cat_cols = ["city" , "route_type" ,"noise_factor"]

#2.One-Hot Encode the train & test Datasets
train = pd.get_dummies(train, columns=cat_cols)
test = pd.get_dummies(test, columns=cat_cols)

#3.Allign test columns with train columns
test = test.reindex(columns=train.columns.drop('peak'), fill_value=0)
print("Train Columns:",train.columns)
print("Test Columns:",test.columns)
```

```python
from sklearn.preprocessing import StandardScaler
#Identify the target numerical column exclude the  target column peak
numeric_col = train.select_dtypes(include='number').columns.drop('peak')
print("Numerical columns to scale:",numeric_col)
```

```python
Scaler = StandardScaler()
#Fit in train numeric cols
Scaler.fit(train[numeric_col])

#Transform Test numeric cols
train[numeric_col] = Scaler.transform(train[numeric_col])
```

```python
train[numeric_col].describe()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import f1_score, classification_report , confusion_matrix
```

```python
#Split Data into features and Target
#Features
X=train.drop('peak',axis=1)

#Target
Y=train['peak']
```

```python
#Train Validation Data
#We split the data so we can evaluate the model before submitting
X_train , X_val , Y_train , Y_val = train_test_split(X,Y,test_size=0.2,random_state=42,stratify=Y)
```

```python
#Train The logistic regression
#Create model
lr_model = LogisticRegression(max_iter=1000,random_state=42)

#Fit the model
lr_model.fit(X_train,Y_train)
```

```python
#Evaluate on Validation set
#Predict
y_pred = lr_model.predict(X_val)

#F1 Score
print(f1_score(Y_val,y_pred))

test = test.reindex(columns=train.columns.drop('peak'), fill_value=0)
```
